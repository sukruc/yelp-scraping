```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
PATH = "data_esin/"
file_list = os.listdir(PATH)
file_list[:5]   # display first five items in file list
```

```python
# load 1 json file and explore the structure:
filename = file_list[0] # select first filename
with open(PATH+filename,"rb") as f:
    sample_json = json.load(f)
```

```python
sample_json
```

```python
# explore keys of json file:
sample_json.keys()
```

```python
# required data is stored under "businesses" key
# let's reach it by calling the key
print(type(sample_json["businesses"]))
sample_json["businesses"]
```

```python
# now we define a for loop to append all json files into one large list
# since target data is a list, we will use extend method to get the element of the list
data = [] # initiate an empty list to store data
for filename in os.listdir(PATH):
    with open(PATH+filename,"rb") as f:
        ceyson = json.load(f)
    data.extend(ceyson["businesses"])
    
```

```python
len(data)
```

```python
# convert data to pandas DataFrame format:

df = pd.DataFrame(data)
df.head() # investigate the first five rows
```

```python
# investigate one row:
df.loc[0]
```

```python
# categories column contains valuable information, so let's take a closer look:
df.categories[0]
```

```python
# We can replace this column to contain only "title"
df.categories = df.categories.apply(lambda x: x[0]["title"] if len(x) > 0 else None)
```

```python
df.categories.head()
```

```python
# Filter dataframe only on Gas Stations:
df[df.categories == "Gas Stations"].head()
```

```python
# apply a similar operation to "coordinates":
df["latitude"] = df.coordinates.apply(lambda x: x["latitude"]) # create a new column "latitude" and extract latitude from oordinates
df["longitude"] = df.coordinates.apply(lambda x: x["longitude"]) # create a new column "longitude" and extract latitude from oordinates
```

```python
df.head()
```

```python
# drop "coordinates" columns as we no longer need it
df.drop("coordinates",axis=1,inplace=True)
```

```python
# we will add a new column: Google maps link
# sample url will look like this: https://www.google.com/maps/@38.938309,-122.9259576,14z
lat = df.latitude.astype(str)
long = df.longitude.astype(str)
zoom = "18z"
df["google_map_link"] = "https://www.google.com/maps/@"+lat+","+long+","+zoom
```

```python
df.google_map_link[0]
```

```python
df["location"][0]
```

```python
df["address1"] = df.location.apply(lambda x: x["address1"])
df["address2"] = df.location.apply(lambda x: x["address2"])
df["address3"] = df.location.apply(lambda x: x["address3"])
df["city"] = df.location.apply(lambda x: x["city"])
df["zip_code"] = df.location.apply(lambda x: x["zip_code"])
df["country"] = df.location.apply(lambda x: x["country"])
df["state"] = df.location.apply(lambda x: x["state"])
df["display_address"] = df.location.apply(lambda x: str(x["display_address"]).strip("[]'").replace("'",""))
```

```python
# clean out unnecessary columns:
unc_cols = ["location"]
df_clean = df.drop(unc_cols,axis=1)
```

```python
# drop duplicate values
non_dup = df_clean.id.drop_duplicates().index.values
df_clean = df_clean.loc[non_dup]
```

```python
df_clean.to_csv("ca_businesses.csv",index=False)
```
